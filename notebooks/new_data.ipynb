```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import plotly.subplots as sp
from plotly.subplots import make_subplots
from scipy import stats
import sklearn

import warnings
warnings.filterwarnings('ignore')
```

```python
# Преобразование числовых колонок
df['Discovery_CPU Count'] = pd.to_numeric(df['Discovery_CPU Count'], errors='coerce')
df['Discovery_RAM (Gb)'] = pd.to_numeric(df['Discovery_RAM (Gb)'], errors='coerce')

# Проверка данных
print("Общая информация о данных:")
print(f"Количество виртуальных машин: {len(df)}")
print(f"Количество уникальных систем: {df['Объект обслуживания (АС/ПС)'].nunique()}")
print(f"Количество хостов vSphere: {df['Discovery_vSphere'].nunique()}")
print("\n" + "="*80)

# Статистика по ресурсам
print("\nСтатистика по CPU:")
print(df['Discovery_CPU Count'].describe())
print("\nСтатистика по RAM (GB):")
print(df['Discovery_RAM (Gb)'].describe())

# Анализ по системам
system_stats = df.groupby('Объект обслуживания (АС/ПС)').agg({
    'Key': 'count',
    'Discovery_CPU Count': 'sum',
    'Discovery_RAM (Gb)': 'sum'
}).rename(columns={'Key': 'Количество ВМ'}).sort_values('Количество ВМ', ascending=False)

print("\nТоп-10 систем по количеству ВМ:")
print(system_stats.head(10))

# Создание интерактивного дашборда
fig = make_subplots(
    rows=4, cols=3,
    subplot_titles=(
        '1. Распределение CPU', '2. Распределение RAM', '3. ВМ по системам (топ-10)',
        '4. Соотношение CPU/RAM', '5. Топ систем по CPU', '6. Топ систем по RAM',
        '7. Плотность CPU', '8. Плотность RAM', '9. ВМ по хостам vSphere',
        '10. CPU vs RAM scatter', '11. Суммарные ресурсы по системам', '12. Heatmap ресурсов'
    ),
    specs=[
        [{"type": "histogram"}, {"type": "histogram"}, {"type": "bar"}],
        [{"type": "scatter"}, {"type": "bar"}, {"type": "bar"}],
        [{"type": "box"}, {"type": "box"}, {"type": "pie"}],
        [{"type": "scatter"}, {"type": "bar"}, {"type": "heatmap"}]
    ],
    vertical_spacing=0.08,
    horizontal_spacing=0.08
)

# 1. Распределение CPU
fig.add_trace(
    go.Histogram(
        x=df['Discovery_CPU Count'],
        nbinsx=30,
        name='CPU распределение',
        marker_color='blue',
        opacity=0.7
    ),
    row=1, col=1
)

# 2. Распределение RAM
fig.add_trace(
    go.Histogram(
        x=df['Discovery_RAM (Gb)'],
        nbinsx=30,
        name='RAM распределение',
        marker_color='green',
        opacity=0.7
    ),
    row=1, col=2
)

# 3. Топ-10 систем по количеству ВМ
top_systems = system_stats.head(10)
fig.add_trace(
    go.Bar(
        x=top_systems.index,
        y=top_systems['Количество ВМ'],
        name='Количество ВМ',
        marker_color='purple',
        text=top_systems['Количество ВМ'],
        textposition='auto'
    ),
    row=1, col=3
)

# 4. Соотношение CPU/RAM
fig.add_trace(
    go.Scatter(
        x=df['Discovery_CPU Count'],
        y=df['Discovery_RAM (Gb)'],
        mode='markers',
        marker=dict(
            size=8,
            color=df['Discovery_CPU Count'],
            colorscale='Viridis',
            showscale=True,
            colorbar=dict(title="CPU ядер")
        ),
        text=df['Имя КЕ'],
        name='CPU vs RAM'
    ),
    row=2, col=1
)

# 5. Топ систем по общему CPU
top_cpu_systems = df.groupby('Объект обслуживания (АС/ПС)')['Discovery_CPU Count'].sum().nlargest(10)
fig.add_trace(
    go.Bar(
        x=top_cpu_systems.values,
        y=top_cpu_systems.index,
        orientation='h',
        name='Суммарный CPU',
        marker_color='orange',
        text=top_cpu_systems.values,
        textposition='auto'
    ),
    row=2, col=2
)

# 6. Топ систем по общему RAM
top_ram_systems = df.groupby('Объект обслуживания (АС/ПС)')['Discovery_RAM (Gb)'].sum().nlargest(10)
fig.add_trace(
    go.Bar(
        x=top_ram_systems.values,
        y=top_ram_systems.index,
        orientation='h',
        name='Суммарный RAM',
        marker_color='red',
        text=top_ram_systems.values,
        textposition='auto'
    ),
    row=2, col=3
)

# 7. Box plot CPU
fig.add_trace(
    go.Box(
        y=df['Discovery_CPU Count'],
        name='CPU распределение',
        boxpoints='outliers',
        marker_color='blue'
    ),
    row=3, col=1
)

# 8. Box plot RAM
fig.add_trace(
    go.Box(
        y=df['Discovery_RAM (Gb)'],
        name='RAM распределение',
        boxpoints='outliers',
        marker_color='green'
    ),
    row=3, col=2
)

# 9. Распределение ВМ по хостам vSphere
vsphere_counts = df['Discovery_vSphere'].value_counts()
fig.add_trace(
    go.Pie(
        labels=vsphere_counts.index,
        values=vsphere_counts.values,
        name='ВМ по vSphere',
        hole=0.4,
        textinfo='percent+label'
    ),
    row=3, col=3
)

# 10. Scatter с размером по CPU
fig.add_trace(
    go.Scatter(
        x=df['Discovery_CPU Count'],
        y=df['Discovery_RAM (Gb)'],
        mode='markers',
        marker=dict(
            size=df['Discovery_RAM (Gb)']/5,
            color=df['Discovery_CPU Count'],
            colorscale='Rainbow',
            showscale=True,
            colorbar=dict(title="CPU ядер"),
            sizemode='area'
        ),
        text=df['Имя КЕ'] + '<br>' + df['Объект обслуживания (АС/ПС)'],
        hoverinfo='text',
        name='Размер = RAM'
    ),
    row=4, col=1
)

# 11. Суммарные ресурсы по системам (топ-8)
top_8_systems = system_stats.head(8)
fig.add_trace(
    go.Bar(
        x=top_8_systems.index,
        y=top_8_systems['Discovery_CPU Count'],
        name='CPU суммарно',
        marker_color='blue',
        text=top_8_systems['Discovery_CPU Count'],
        textposition='auto'
    ),
    row=4, col=2
)

fig.add_trace(
    go.Bar(
        x=top_8_systems.index,
        y=top_8_systems['Discovery_RAM (Gb)'],
        name='RAM суммарно',
        marker_color='green',
        text=top_8_systems['Discovery_RAM (Gb)'],
        textposition='auto',
        visible=False
    ),
    row=4, col=2
)

# Добавляем кнопки для переключения
fig.update_layout(
    updatemenus=[
        dict(
            type="buttons",
            direction="left",
            buttons=list([
                dict(
                    args=[{"visible": [True, False]},
                          {"title": "Суммарный CPU по системам (топ-8)"}],
                    label="Показать CPU",
                    method="update"
                ),
                dict(
                    args=[{"visible": [False, True]},
                          {"title": "Суммарный RAM по системам (топ-8)"}],
                    label="Показать RAM",
                    method="update"
                ),
            ]),
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.5,
            xanchor="center",
            y=1.15,
            yanchor="top"
        ),
    ]
)

# 12. Heatmap распределения ресурсов
# Создаем матрицу для heatmap
resource_matrix = df.pivot_table(
    values='Key',
    index='Объект обслуживания (АС/ПС)',
    columns=pd.cut(df['Discovery_CPU Count'], bins=[0, 4, 8, 16, 32, 100]),
    aggfunc='count',
    fill_value=0
)

# Берем только топ-15 систем для heatmap
top_15_systems = system_stats.head(15).index
heatmap_data = resource_matrix.loc[resource_matrix.index.isin(top_15_systems)]

fig.add_trace(
    go.Heatmap(
        z=heatmap_data.values,
        x=[str(col) for col in heatmap_data.columns],
        y=heatmap_data.index,
        colorscale='YlOrRd',
        showscale=True,
        name='Heatmap ресурсов'
    ),
    row=4, col=3
)

# Обновление макета
fig.update_layout(
    title_text='Дашборд анализа виртуальных машин',
    height=1600,
    width=1400,
    showlegend=False,
    template='plotly_white'
)

# Обновление подписей осей
fig.update_xaxes(title_text="Количество CPU ядер", row=1, col=1)
fig.update_xaxes(title_text="RAM (GB)", row=1, col=2)
fig.update_xaxes(title_text="Системы", row=1, col=3)
fig.update_xaxes(title_text="CPU ядер", row=2, col=1)
fig.update_yaxes(title_text="RAM (GB)", row=2, col=1)
fig.update_xaxes(title_text="Суммарный CPU", row=2, col=2)
fig.update_xaxes(title_text="Суммарный RAM (GB)", row=2, col=3)
fig.update_yaxes(title_text="CPU ядер", row=3, col=1)
fig.update_yaxes(title_text="RAM (GB)", row=3, col=2)
fig.update_xaxes(title_text="CPU ядер", row=4, col=1)
fig.update_yaxes(title_text="RAM (GB)", row=4, col=1)
fig.update_xaxes(title_text="Системы", row=4, col=2)
fig.update_yaxes(title_text="Ресурсы", row=4, col=2)
fig.update_xaxes(title_text="Диапазон CPU", row=4, col=3)
fig.update_yaxes(title_text="Системы", row=4, col=3)

# Поворачиваем подписи для лучшей читаемости
fig.update_xaxes(tickangle=45, row=1, col=3)
fig.update_xaxes(tickangle=45, row=4, col=2)

# Сохраняем дашборд
fig.write_html("vm_analysis_dashboard.html")

print("\n" + "="*80)
print("АНАЛИЗ ДАННЫХ:")
print("="*80)

# Подробный анализ
print("\n1. ОБЩАЯ СТАТИСТИКА:")
print(f"   • Всего виртуальных машин: {len(df)}")
print(f"   • Среднее количество CPU: {df['Discovery_CPU Count'].mean():.1f} ядер")
print(f"   • Средний объем RAM: {df['Discovery_RAM (Gb)'].mean():.1f} GB")
print(f"   • Медиана CPU: {df['Discovery_CPU Count'].median()} ядер")
print(f"   • Медиана RAM: {df['Discovery_RAM (Gb)'].median()} GB")

print("\n2. РАСПРЕДЕЛЕНИЕ ПО СИСТЕМАМ:")
for i, (system, count) in enumerate(system_stats['Количество ВМ'].head(5).items(), 1):
    print(f"   {i}. {system}: {count} ВМ")

print("\n3. РАСПРЕДЕЛЕНИЕ ПО РЕСУРСАМ:")
cpu_bins = pd.cut(df['Discovery_CPU Count'], bins=[0, 2, 4, 8, 16, 32, 100])
ram_bins = pd.cut(df['Discovery_RAM (Gb)'], bins=[0, 4, 8, 16, 32, 64, 128, 200])

print("   CPU распределение:")
for bin_name, count in cpu_bins.value_counts().sort_index().items():
    print(f"     • {bin_name}: {count} ВМ ({count/len(df)*100:.1f}%)")

print("\n   RAM распределение:")
for bin_name, count in ram_bins.value_counts().sort_index().items():
    print(f"     • {bin_name}: {count} ВМ ({count/len(df)*100:.1f}%)")

print("\n4. ВМ С МАКСИМАЛЬНЫМИ РЕСУРСАМИ:")
max_cpu_vm = df.loc[df['Discovery_CPU Count'].idxmax()]
max_ram_vm = df.loc[df['Discovery_RAM (Gb)'].idxmax()]
print(f"   • Макс CPU: {max_cpu_vm['Имя КЕ']} - {max_cpu_vm['Discovery_CPU Count']} ядер ({max_cpu_vm['Объект обслуживания (АС/ПС)']})")
print(f"   • Макс RAM: {max_ram_vm['Имя КЕ']} - {max_ram_vm['Discovery_RAM (Gb)']} GB ({max_ram_vm['Объект обслуживания (АС/ПС)']})")

print("\n5. АНАЛИЗ vSPHERE ХОСТОВ:")
print(f"   • Основной хост: vcsa67.belpsb.by - {len(df[df['Discovery_vSphere'] == 'vcsa67.belpsb.by'])} ВМ")
print(f"   • Резервный хост: vcsa70-old - {len(df[df['Discovery_vSphere'] == 'vcsa70-old'])} ВМ")

print("\n6. РЕКОМЕНДАЦИИ ПО ОПТИМИЗАЦИИ:")
print("   • Рассмотреть возможность консолидации ВМ с малыми ресурсами (2 CPU/4GB RAM)")
print("   • Проанализировать ВМ с большим объемом RAM (>64GB) на предмет оптимизации")
print("   • Проверить балансировку ВМ между хостами vSphere")
print("   • Оценить возможность использования контейнеризации для микросервисов")

# Выводим ссылку на дашборд
print("\n" + "="*80)
print(f"✅ Дашборд сохранен как 'vm_analysis_dashboard.html'")
print("Откройте файл в браузере для просмотра интерактивных графиков")
print("="*80)
```

```python
# Преобразование числовых колонок
df['Discovery_CPU Count'] = pd.to_numeric(df['Discovery_CPU Count'], errors='coerce')
df['Discovery_RAM (Gb)'] = pd.to_numeric(df['Discovery_RAM (Gb)'], errors='coerce')

# Создаем категории для ресурсов
df['CPU_Category'] = pd.cut(df['Discovery_CPU Count'],
                            bins=[0, 2, 4, 8, 16, 32, 100],
                            labels=['1-2', '3-4', '5-8', '9-16', '17-32', '33+'])

df['RAM_Category'] = pd.cut(df['Discovery_RAM (Gb)'],
                           bins=[0, 4, 8, 16, 32, 64, 128, 200],
                           labels=['1-4', '5-8', '9-16', '17-32', '33-64', '65-128', '129+'])

# Вычисляем корреляцию
correlation = df['Discovery_CPU Count'].corr(df['Discovery_RAM (Gb)'])

print("="*80)
print("АНАЛИЗ ВЗАИМОСВЯЗЕЙ CPU И RAM")
print("="*80)
print(f"Корреляция между CPU и RAM: {correlation:.3f}")

# Статистический анализ по системам
system_analysis = df.groupby('Объект обслуживания (АС/ПС)').agg({
    'Key': 'count',
    'Discovery_CPU Count': ['mean', 'median', 'std', 'sum'],
    'Discovery_RAM (Gb)': ['mean', 'median', 'std', 'sum']
})

# Переименовываем колонки
system_analysis.columns = ['Количество_ВМ', 'CPU_среднее', 'CPU_медиана', 'CPU_std', 'CPU_сумма',
                          'RAM_среднее', 'RAM_медиана', 'RAM_std', 'RAM_сумма']
system_analysis = system_analysis.sort_values('Количество_ВМ', ascending=False)

print("\nСТАТИСТИКА ПО СИСТЕМАМ (топ-5):")
print(system_analysis.head())

# Создаем дашборд для анализа взаимосвязей
fig = make_subplots(
    rows=4, cols=3,
    subplot_titles=(
        '1. Scatter: CPU vs RAM с цветом по системе',
        '2. Scatter с размером по RAM (логарифм)',
        '3. Heatmap: Совместное распределение CPU/RAM',
        '4. Boxplot: CPU по категориям RAM',
        '5. Boxplot: RAM по категориям CPU',
        '6. 2D Histogram: Плотность совместного распределения',
        '7. Категоризированный scatter',
        '8. Тепловая карта систем',
        '9. Корреляционная матрица',
        '10. Распределение CPU/RAM для топ-5 систем',
        '11. Взаимосвязь CPU/RAM по хостам vSphere',
        '12. Анализ выбросов CPU/RAM'
    ),
    specs=[
        [{"type": "scatter"}, {"type": "scatter"}, {"type": "heatmap"}],
        [{"type": "box"}, {"type": "box"}, {"type": "histogram2d"}],
        [{"type": "scatter"}, {"type": "heatmap"}, {"type": "heatmap"}],
        [{"type": "scatter"}, {"type": "scatter"}, {"type": "scatter"}]
    ],
    vertical_spacing=0.08,
    horizontal_spacing=0.08
)

# 1. Scatter: CPU vs RAM с цветом по системе (топ-8 систем)
top_systems = df['Объект обслуживания (АС/ПС)'].value_counts().head(8).index.tolist()
df_top = df[df['Объект обслуживания (АС/ПС)'].isin(top_systems)]

fig.add_trace(
    go.Scatter(
        x=df_top['Discovery_CPU Count'],
        y=df_top['Discovery_RAM (Gb)'],
        mode='markers',
        marker=dict(
            size=10,
            color=pd.Categorical(df_top['Объект обслуживания (АС/ПС)']).codes,
            colorscale='Viridis',
            showscale=True,
            colorbar=dict(title="Системы", tickvals=list(range(len(top_systems))),
                         ticktext=top_systems)
        ),
        text=df_top['Имя КЕ'] + '<br>' + df_top['Объект обслуживания (АС/ПС)'],
        hoverinfo='text',
        name='CPU vs RAM'
    ),
    row=1, col=1
)

# Добавляем линию регрессии для графика 1
x_range = np.linspace(df['Discovery_CPU Count'].min(), df['Discovery_CPU Count'].max(), 100)
z = np.polyfit(df['Discovery_CPU Count'], df['Discovery_RAM (Gb)'], 1)
p = np.poly1d(z)

fig.add_trace(
    go.Scatter(
        x=x_range,
        y=p(x_range),
        mode='lines',
        line=dict(color='red', width=2, dash='dash'),
        name=f'Линия регрессии (корр: {correlation:.3f})'
    ),
    row=1, col=1
)

# 2. Scatter с размером по RAM (логарифмическая шкала)
fig.add_trace(
    go.Scatter(
        x=df['Discovery_CPU Count'],
        y=df['Discovery_RAM (Gb)'],
        mode='markers',
        marker=dict(
            size=np.log10(df['Discovery_RAM (Gb)'] + 1) * 8,
            color=np.log10(df['Discovery_CPU Count'] + 1),
            colorscale='Rainbow',
            showscale=True,
            colorbar=dict(title="log(CPU)")
        ),
        text=df['Имя КЕ'] + '<br>CPU: ' + df['Discovery_CPU Count'].astype(str) +
             ', RAM: ' + df['Discovery_RAM (Gb)'].astype(str) + 'GB',
        hoverinfo='text',
        name='Размер = log(RAM)'
    ),
    row=1, col=2
)

# 3. Heatmap: Совместное распределение CPU/RAM
heatmap_data, xedges, yedges = np.histogram2d(
    df['Discovery_CPU Count'],
    df['Discovery_RAM (Gb)'],
    bins=[15, 15]
)

fig.add_trace(
    go.Heatmap(
        z=heatmap_data.T,
        x=xedges,
        y=yedges,
        colorscale='YlOrRd',
        showscale=True,
        name='Совместное распределение'
    ),
    row=1, col=3
)

# 4. Boxplot: CPU по категориям RAM
fig.add_trace(
    go.Box(
        x=df['RAM_Category'],
        y=df['Discovery_CPU Count'],
        boxpoints='all',
        pointpos=-1.5,
        jitter=0.3,
        marker=dict(size=5, opacity=0.5),
        name='CPU по RAM категориям'
    ),
    row=2, col=1
)

# 5. Boxplot: RAM по категориям CPU
fig.add_trace(
    go.Box(
        x=df['CPU_Category'],
        y=df['Discovery_RAM (Gb)'],
        boxpoints='all',
        pointpos=-1.5,
        jitter=0.3,
        marker=dict(size=5, opacity=0.5),
        name='RAM по CPU категориям'
    ),
    row=2, col=2
)

# 6. 2D Histogram: Плотность совместного распределения
fig.add_trace(
    go.Histogram2dContour(
        x=df['Discovery_CPU Count'],
        y=df['Discovery_RAM (Gb)'],
        colorscale='Hot',
        contours=dict(coloring='heatmap'),
        name='Плотность распределения'
    ),
    row=2, col=3
)

# 7. Категоризированный scatter по CPU категориям
for i, cpu_cat in enumerate(df['CPU_Category'].cat.categories):
    df_cat = df[df['CPU_Category'] == cpu_cat]
    fig.add_trace(
        go.Scatter(
            x=df_cat['Discovery_CPU Count'],
            y=df_cat['Discovery_RAM (Gb)'],
            mode='markers',
            marker=dict(size=8),
            name=f'CPU: {cpu_cat}',
            visible=(i == 0)  # Показываем только первую категорию по умолчанию
        ),
        row=3, col=1
    )

# Добавляем кнопки для переключения CPU категорий
buttons = []
for i, cpu_cat in enumerate(df['CPU_Category'].cat.categories):
    visible = [False] * len(df['CPU_Category'].cat.categories)
    visible[i] = True
    buttons.append(
        dict(
            label=f'CPU: {cpu_cat}',
            method='update',
            args=[{'visible': visible * 7},  # Умножаем на количество графиков в ряду
                  {'title': f'Категория CPU: {cpu_cat}'}]
        )
    )

fig.update_layout(
    updatemenus=[
        dict(
            type="buttons",
            direction="right",
            buttons=buttons,
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.1,
            xanchor="left",
            y=1.15,
            yanchor="top"
        ),
    ]
)

# 8. Тепловая карта систем (средние значения)
system_heatmap = df.groupby('Объект обслуживания (АС/ПС)').agg({
    'Discovery_CPU Count': 'mean',
    'Discovery_RAM (Gb)': 'mean'
}).round(1)

# Берем топ-15 систем
system_heatmap = system_heatmap.head(15)

fig.add_trace(
    go.Heatmap(
        z=system_heatmap.values.T,
        x=system_heatmap.index,
        y=['Средний CPU', 'Средний RAM (GB)'],
        colorscale='RdYlGn',
        text=system_heatmap.values.T,
        texttemplate='%{text}',
        textfont={"size": 10},
        showscale=True,
        name='Средние значения по системам'
    ),
    row=3, col=2
)

# 9. Корреляционная матрица для числовых данных
corr_matrix = df[['Discovery_CPU Count', 'Discovery_RAM (Gb)']].corr()

fig.add_trace(
    go.Heatmap(
        z=corr_matrix.values,
        x=corr_matrix.columns,
        y=corr_matrix.columns,
        colorscale='RdBu',
        zmin=-1, zmax=1,
        text=corr_matrix.values.round(3),
        texttemplate='%{text}',
        textfont={"size": 14},
        showscale=True,
        name='Корреляционная матрица'
    ),
    row=3, col=3
)

# 10. Распределение CPU/RAM для топ-5 систем
top_5_systems = system_analysis.head(5).index
for i, system in enumerate(top_5_systems):
    df_system = df[df['Объект обслуживания (АС/ПС)'] == system]

    # CPU распределение
    fig.add_trace(
        go.Violin(
            x=[system] * len(df_system),
            y=df_system['Discovery_CPU Count'],
            name=f'{system} CPU',
            side='positive',
            line_color=px.colors.qualitative.Set1[i],
            opacity=0.7,
            showlegend=False,
            visible=(i == 0)  # Показываем только первую систему
        ),
        row=4, col=1
    )

    # RAM распределение
    fig.add_trace(
        go.Violin(
            x=[system] * len(df_system),
            y=df_system['Discovery_RAM (Gb)'],
            name=f'{system} RAM',
            side='negative',
            line_color=px.colors.qualitative.Set2[i],
            opacity=0.7,
            showlegend=False,
            visible=(i == 0)  # Показываем только первую систему
        ),
        row=4, col=1
    )

# 11. Взаимосвязь CPU/RAM по хостам vSphere
vsphere_hosts = df['Discovery_vSphere'].value_counts().index.tolist()

for i, host in enumerate(vsphere_hosts[:3]):  # Берем топ-3 хоста
    df_host = df[df['Discovery_vSphere'] == host]

    fig.add_trace(
        go.Scatter(
            x=df_host['Discovery_CPU Count'],
            y=df_host['Discovery_RAM (Gb)'],
            mode='markers',
            marker=dict(size=10, symbol=i+1),  # Разные символы для разных хостов
            name=host,
            text=df_host['Имя КЕ'],
            visible=(i == 0)  # Показываем только первый хост
        ),
        row=4, col=2
    )

# 12. Анализ выбросов CPU/RAM
# Определяем выбросы с помощью метода IQR
Q1_cpu = df['Discovery_CPU Count'].quantile(0.25)
Q3_cpu = df['Discovery_CPU Count'].quantile(0.75)
IQR_cpu = Q3_cpu - Q1_cpu

Q1_ram = df['Discovery_RAM (Gb)'].quantile(0.25)
Q3_ram = df['Discovery_RAM (Gb)'].quantile(0.75)
IQR_ram = Q3_ram - Q1_ram

# Метки выбросов
df['CPU_Outlier'] = (df['Discovery_CPU Count'] < (Q1_cpu - 1.5 * IQR_cpu)) | \
                    (df['Discovery_CPU Count'] > (Q3_cpu + 1.5 * IQR_cpu))
df['RAM_Outlier'] = (df['Discovery_RAM (Gb)'] < (Q1_ram - 1.5 * IQR_ram)) | \
                    (df['Discovery_RAM (Gb)'] > (Q3_ram + 1.5 * IQR_ram))

fig.add_trace(
    go.Scatter(
        x=df['Discovery_CPU Count'],
        y=df['Discovery_RAM (Gb)'],
        mode='markers',
        marker=dict(
            size=8,
            color=np.where(df['CPU_Outlier'] | df['RAM_Outlier'], 'red', 'blue'),
            opacity=0.7
        ),
        text=df['Имя КЕ'] + '<br>' +
             np.where(df['CPU_Outlier'] | df['RAM_Outlier'],
                     'ВЫБРОС', 'Норма') + '<br>' +
             'CPU: ' + df['Discovery_CPU Count'].astype(str) + '<br>' +
             'RAM: ' + df['Discovery_RAM (Gb)'].astype(str) + 'GB',
        hoverinfo='text',
        name='Анализ выбросов'
    ),
    row=4, col=3
)

# Добавляем линии для выбросов
fig.add_trace(
    go.Scatter(
        x=[Q3_cpu + 1.5 * IQR_cpu, Q3_cpu + 1.5 * IQR_cpu],
        y=[df['Discovery_RAM (Gb)'].min(), df['Discovery_RAM (Gb)'].max()],
        mode='lines',
        line=dict(color='orange', width=1, dash='dash'),
        name='Граница CPU выбросов',
        showlegend=False
    ),
    row=4, col=3
)

fig.add_trace(
    go.Scatter(
        x=[df['Discovery_CPU Count'].min(), df['Discovery_CPU Count'].max()],
        y=[Q3_ram + 1.5 * IQR_ram, Q3_ram + 1.5 * IQR_ram],
        mode='lines',
        line=dict(color='orange', width=1, dash='dash'),
        name='Граница RAM выбросов',
        showlegend=False
    ),
    row=4, col=3
)

# Обновление макета
fig.update_layout(
    title_text='АНАЛИЗ ВЗАИМОСВЯЗЕЙ CPU И RAM ВИРТУАЛЬНЫХ СЕРВЕРОВ',
    height=1600,
    width=1600,
    showlegend=True,
    template='plotly_white'
)

# Обновление подписей осей
fig.update_xaxes(title_text="CPU ядра", row=1, col=1)
fig.update_yaxes(title_text="RAM (GB)", row=1, col=1)
fig.update_xaxes(title_text="CPU ядра", row=1, col=2)
fig.update_yaxes(title_text="RAM (GB)", row=1, col=2)
fig.update_xaxes(title_text="CPU ядра", row=1, col=3)
fig.update_yaxes(title_text="RAM (GB)", row=1, col=3)
fig.update_xaxes(title_text="RAM категории", row=2, col=1)
fig.update_yaxes(title_text="CPU ядра", row=2, col=1)
fig.update_xaxes(title_text="CPU категории", row=2, col=2)
fig.update_yaxes(title_text="RAM (GB)", row=2, col=2)
fig.update_xaxes(title_text="CPU ядра", row=2, col=3)
fig.update_yaxes(title_text="RAM (GB)", row=2, col=3)
fig.update_xaxes(title_text="CPU ядра", row=3, col=1)
fig.update_yaxes(title_text="RAM (GB)", row=3, col=1)
fig.update_xaxes(title_text="Системы", row=3, col=2)
fig.update_yaxes(title_text="Метрики", row=3, col=2)
fig.update_xaxes(title_text="Параметры", row=3, col=3)
fig.update_yaxes(title_text="Параметры", row=3, col=3)
fig.update_xaxes(title_text="Системы", row=4, col=1)
fig.update_yaxes(title_text="Ресурсы", row=4, col=1)
fig.update_xaxes(title_text="CPU ядра", row=4, col=2)
fig.update_yaxes(title_text="RAM (GB)", row=4, col=2)
fig.update_xaxes(title_text="CPU ядра", row=4, col=3)
fig.update_yaxes(title_text="RAM (GB)", row=4, col=3)

# Поворачиваем подписи для лучшей читаемости
fig.update_xaxes(tickangle=45, row=1, col=1)
fig.update_xaxes(tickangle=45, row=3, col=2)

# Сохраняем дашборд
fig.write_html("cpu_ram_relationships_dashboard.html")

print("\n" + "="*80)
print("КЛЮЧЕВЫЕ ВЫВОДЫ:")
print("="*80)

# Анализ распределений
print("\n1. РАСПРЕДЕЛЕНИЕ ПО КАТЕГОРИЯМ:")

cpu_dist = df['CPU_Category'].value_counts().sort_index()
print("\n   Распределение по CPU:")
for cat, count in cpu_dist.items():
    percentage = (count / len(df)) * 100
    print(f"     • {cat} ядер: {count} ВМ ({percentage:.1f}%)")

ram_dist = df['RAM_Category'].value_counts().sort_index()
print("\n   Распределение по RAM:")
for cat, count in ram_dist.items():
    percentage = (count / len(df)) * 100
    print(f"     • {cat} GB: {count} ВМ ({percentage:.1f}%)")

# Анализ совместных распределений
print("\n2. СОВМЕСТНЫЕ РАСПРЕДЕЛЕНИЯ:")
joint_counts = pd.crosstab(df['CPU_Category'], df['RAM_Category'])
top_combinations = joint_counts.stack().sort_values(ascending=False).head(5)

print("\n   Самые частые комбинации CPU/RAM:")
for (cpu_cat, ram_cat), count in top_combinations.items():
    percentage = (count / len(df)) * 100
    print(f"     • CPU {cpu_cat}, RAM {ram_cat}: {count} ВМ ({percentage:.1f}%)")

# Анализ выбросов
outliers_cpu = df[df['CPU_Outlier']]
outliers_ram = df[df['RAM_Outlier']]

print(f"\n3. АНАЛИЗ ВЫБРОСОВ:")
print(f"   • Выбросы по CPU: {len(outliers_cpu)} ВМ")
print(f"   • Выбросы по RAM: {len(outliers_ram)} ВМ")
print(f"   • Общие выбросы: {len(df[df['CPU_Outlier'] & df['RAM_Outlier']])} ВМ")

if len(outliers_cpu) > 0:
    print("\n   Топ выбросов по CPU:")
    for _, row in outliers_cpu.nlargest(3, 'Discovery_CPU Count').iterrows():
        print(f"     • {row['Имя КЕ']}: {row['Discovery_CPU Count']} CPU, {row['Discovery_RAM (Gb)']} GB RAM")

if len(outliers_ram) > 0:
    print("\n   Топ выбросов по RAM:")
    for _, row in outliers_ram.nlargest(3, 'Discovery_RAM (Gb)').iterrows():
        print(f"     • {row['Имя КЕ']}: {row['Discovery_CPU Count']} CPU, {row['Discovery_RAM (Gb)']} GB RAM")

# Анализ по системам
print("\n4. ХАРАКТЕРИСТИКИ ПО СИСТЕМАМ:")

print("\n   Системы с наибольшим средним CPU:")
for system, data in system_analysis.nlargest(3, 'CPU_среднее').iterrows():
    print(f"     • {system}: {data['CPU_среднее']:.1f} CPU, {data['CPU_сумма']} CPU всего")

print("\n   Системы с наибольшим средним RAM:")
for system, data in system_analysis.nlargest(3, 'RAM_среднее').iterrows():
    print(f"     • {system}: {data['RAM_среднее']:.1f} GB RAM, {data['RAM_сумма']} GB всего")

# Анализ баланса ресурсов
df['CPU_RAM_Ratio'] = df['Discovery_CPU Count'] / df['Discovery_RAM (Gb)']
print(f"\n5. БАЛАНС РЕСУРСОВ (CPU/RAM Ratio):")
print(f"   • Среднее соотношение: {df['CPU_RAM_Ratio'].mean():.3f}")
print(f"   • Медиана соотношения: {df['CPU_RAM_Ratio'].median():.3f}")

print("\n   ВМ с дисбалансом ресурсов (Ratio < 0.1 или > 1.0):")
imbalanced = df[(df['CPU_RAM_Ratio'] < 0.1) | (df['CPU_RAM_Ratio'] > 1.0)].sort_values('CPU_RAM_Ratio')
for _, row in imbalanced.head(5).iterrows():
    print(f"     • {row['Имя КЕ']}: {row['CPU_RAM_Ratio']:.3f} (CPU: {row['Discovery_CPU Count']}, RAM: {row['Discovery_RAM (Gb)']}GB)")

print("\n" + "="*80)
print(f"✅ Дашборд сохранен как 'cpu_ram_relationships_dashboard.html'")
print("Откройте файл в браузере для просмотра интерактивных графиков")
print("="*80)

# Дополнительный анализ: кластеризация по ресурсам
from sklearn.cluster import KMeans

print("\n6. КЛАСТЕРИЗАЦИЯ ВМ ПО РЕСУРСАМ:")
# Нормализуем данные
X = df[['Discovery_CPU Count', 'Discovery_RAM (Gb)']].values
X_norm = (X - X.mean(axis=0)) / X.std(axis=0)

# Применяем K-means
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
df['Resource_Cluster'] = kmeans.fit_predict(X_norm)

# Анализируем кластеры
cluster_stats = df.groupby('Resource_Cluster').agg({
    'Discovery_CPU Count': ['mean', 'std', 'min', 'max'],
    'Discovery_RAM (Gb)': ['mean', 'std', 'min', 'max'],
    'Key': 'count'
})

print("\n   Характеристики кластеров:")
for cluster in range(5):
    cluster_data = df[df['Resource_Cluster'] == cluster]
    print(f"\n   Кластер {cluster}: {len(cluster_data)} ВМ")
    print(f"     • CPU: {cluster_data['Discovery_CPU Count'].mean():.1f} ± {cluster_data['Discovery_CPU Count'].std():.1f}")
    print(f"     • RAM: {cluster_data['Discovery_RAM (Gb)'].mean():.1f} ± {cluster_data['Discovery_RAM (Gb)'].std():.1f}")

    # Наиболее частые системы в кластере
    top_systems = cluster_data['Объект обслуживания (АС/ПС)'].value_counts().head(3)
    if len(top_systems) > 0:
        print(f"     • Основные системы: {', '.join([f'{s}({c})' for s, c in top_systems.items()])}")

print("\n" + "="*80)
print("РЕКОМЕНДАЦИИ:")
print("="*80)
print("1. Оптимизировать ВМ с дисбалансом CPU/RAM (соотношение < 0.1 или > 1.0)")
print("2. Рассмотреть возможность консолидации ВМ в кластерах с малыми ресурсами")
print("3. Проанализировать ВМ-выбросы на предмет возможности ресайза")
print("4. Учитывать специфику распределения ресурсов при планировании нагрузки")
print("5. Использовать кластерный анализ для группового управления ВМ")
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

# Загрузка данных
df = pd.read_excel('metrics_small.xlsx')

# Преобразование типов данных
df['Timestamp'] = pd.to_datetime(df['Timestamp'], format='%d.%m.%y %H:%M:%S')

# Основная информация
print("="*80)
print("АНАЛИЗ МЕТРИК ВИРТУАЛЬНОГО СЕРВЕРА AIRFLOW1")
print("="*80)

print(f"\nПериод анализа: с {df['Timestamp'].min()} по {df['Timestamp'].max()}")
print(f"Количество измерений: {len(df)}")
print(f"Интервал измерений: 30 минут")

# Статистика по метрикам
print("\nОСНОВНАЯ СТАТИСТИКА ПО МЕТРИКАМ:")
metrics_summary = df.describe().round(2)
print(metrics_summary)

# Нормализация метрик для анализа
metrics_columns = [
    'cpu.ready.summation', 'cpu.usage.average', 'cpu.usagemhz.average',
    'disk.usage.average', 'mem.consumed.average', 'mem.usage.average',
    'mem.vmmemctl.average', 'net.usage.average'
]

# Анализ трендов и паттернов
print("\n" + "="*80)
print("АНАЛИЗ ТРЕНДОВ И ПАТТЕРНОВ")
print("="*80)

# Анализ пиковых нагрузок
peak_usage = {
    'CPU Usage': df['cpu.usage.average'].max(),
    'CPU Ready': df['cpu.ready.summation'].max(),
    'CPU MHz': df['cpu.usagemhz.average'].max(),
    'Disk Usage': df['disk.usage.average'].max(),
    'Memory Usage': df['mem.usage.average'].max(),
    'Network Usage': df['net.usage.average'].max()
}

print("\nПИКОВЫЕ НАГРУЗКИ:")
for metric, value in peak_usage.items():
    print(f"  • {metric}: {value}")

# Времена пиковой нагрузки
peak_times = {}
for metric in ['cpu.usage.average', 'cpu.ready.summation', 'disk.usage.average', 'net.usage.average']:
    peak_idx = df[metric].idxmax()
    peak_times[metric] = df.loc[peak_idx, 'Timestamp']

print("\nВРЕМЕНА ПИКОВОЙ НАГРУЗКИ:")
for metric, time in peak_times.items():
    print(f"  • {metric}: {time}")

# Создание комплексного дашборда
fig = make_subplots(
    rows=4, cols=3,
    subplot_titles=(
        '1. CPU Usage (%) по времени',
        '2. CPU Ready (ms) по времени',
        '3. CPU MHz Usage по времени',
        '4. Disk Usage (KB/s) по времени',
        '5. Memory Usage (%) по времени',
        '6. Network Usage (KB/s) по времени',
        '7. Memory Consumption (MB) по времени',
        '8. Корреляционная матрица метрик',
        '9. Распределение метрик (Boxplot)',
        '10. Тепловая карта нагрузки по часам',
        '11. Анализ паттернов нагрузки CPU',
        '12. Сводный анализ эффективности'
    ),
    specs=[
        [{"type": "scatter"}, {"type": "scatter"}, {"type": "scatter"}],
        [{"type": "scatter"}, {"type": "scatter"}, {"type": "scatter"}],
        [{"type": "scatter"}, {"type": "heatmap"}, {"type": "box"}],
        [{"type": "heatmap"}, {"type": "scatter"}, {"type": "table"}]
    ],
    vertical_spacing=0.08,
    horizontal_spacing=0.08
)

# 1. CPU Usage (%) по времени
fig.add_trace(
    go.Scatter(
        x=df['Timestamp'],
        y=df['cpu.usage.average'],
        mode='lines+markers',
        line=dict(color='red', width=2),
        marker=dict(size=4),
        name='CPU Usage %',
        hovertemplate='%{x}<br>CPU Usage: %{y:.2f}%<extra></extra>'
    ),
    row=1, col=1
)

# Добавляем горизонтальные линии для порогов
fig.add_trace(
    go.Scatter(
        x=[df['Timestamp'].min(), df['Timestamp'].max()],
        y=[70, 70],
        mode='lines',
        line=dict(color='orange', width=1, dash='dash'),
        name='Предупреждение (70%)',
        showlegend=False
    ),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(
        x=[df['Timestamp'].min(), df['Timestamp'].max()],
        y=[85, 85],
        mode='lines',
        line=dict(color='red', width=1, dash='dash'),
        name='Критично (85%)',
        showlegend=False
    ),
    row=1, col=1
)

# 2. CPU Ready (ms) по времени
fig.add_trace(
    go.Scatter(
        x=df['Timestamp'],
        y=df['cpu.ready.summation'],
        mode='lines+markers',
        line=dict(color='blue', width=2),
        marker=dict(size=4),
        name='CPU Ready (ms)',
        hovertemplate='%{x}<br>CPU Ready: %{y:,} ms<extra></extra>'
    ),
    row=1, col=2
)

# Пороги для CPU Ready (рекомендуется < 2000 ms на ядро)
fig.add_trace(
    go.Scatter(
        x=[df['Timestamp'].min(), df['Timestamp'].max()],
        y=[5000, 5000],
        mode='lines',
        line=dict(color='orange', width=1, dash='dash'),
        name='Предупреждение',
        showlegend=False
    ),
    row=1, col=2
)

# 3. CPU MHz Usage по времени
fig.add_trace(
    go.Scatter(
        x=df['Timestamp'],
        y=df['cpu.usagemhz.average'],
        mode='lines+markers',
        line=dict(color='green', width=2),
        marker=dict(size=4),
        name='CPU MHz',
        hovertemplate='%{x}<br>CPU MHz: %{y:,}<extra></extra>'
    ),
    row=1, col=3
)

# 4. Disk Usage (KB/s) по времени
fig.add_trace(
    go.Scatter(
        x=df['Timestamp'],
        y=df['disk.usage.average'],
        mode='lines+markers',
        line=dict(color='purple', width=2),
        marker=dict(size=4),
        name='Disk Usage (KB/s)',
        hovertemplate='%{x}<br>Disk: %{y} KB/s<extra></extra>'
    ),
    row=2, col=1
)

# 5. Memory Usage (%) по времени
fig.add_trace(
    go.Scatter(
        x=df['Timestamp'],
        y=df['mem.usage.average'],
        mode='lines+markers',
        line=dict(color='orange', width=2),
        marker=dict(size=4),
        name='Memory Usage %',
        hovertemplate='%{x}<br>Memory: %{y:.2f}%<extra></extra>'
    ),
    row=2, col=2
)

# 6. Network Usage (KB/s) по времени
fig.add_trace(
    go.Scatter(
        x=df['Timestamp'],
        y=df['net.usage.average'],
        mode='lines+markers',
        line=dict(color='cyan', width=2),
        marker=dict(size=4),
        name='Network (KB/s)',
        hovertemplate='%{x}<br>Network: %{y} KB/s<extra></extra>'
    ),
    row=2, col=3
)

# 7. Memory Consumption (MB) по времени
fig.add_trace(
    go.Scatter(
        x=df['Timestamp'],
        y=df['mem.consumed.average'] / 1024 / 1024,  # Convert to MB
        mode='lines+markers',
        line=dict(color='brown', width=2),
        marker=dict(size=4),
        name='Memory Consumed (MB)',
        hovertemplate='%{x}<br>Memory: %{y:.0f} MB<extra></extra>'
    ),
    row=3, col=1
)

# 8. Корреляционная матрица метрик
correlation_matrix = df[metrics_columns].corr()

fig.add_trace(
    go.Heatmap(
        z=correlation_matrix.values,
        x=correlation_matrix.columns,
        y=correlation_matrix.columns,
        colorscale='RdBu',
        zmin=-1, zmax=1,
        text=correlation_matrix.values.round(2),
        texttemplate='%{text}',
        textfont={"size": 10},
        showscale=True,
        name='Корреляция'
    ),
    row=3, col=2
)

# 9. Распределение метрик (Boxplot)
for i, metric in enumerate(['cpu.usage.average', 'cpu.ready.summation',
                           'disk.usage.average', 'mem.usage.average']):
    fig.add_trace(
        go.Box(
            y=df[metric],
            name=metric.replace('.', ' '),
            boxpoints='outliers',
            marker_color=px.colors.qualitative.Set1[i],
            showlegend=False
        ),
        row=3, col=3
    )

# 10. Тепловая карта нагрузки по часам
df['Hour'] = df['Timestamp'].dt.hour
df['DayPart'] = pd.cut(df['Hour'],
                       bins=[0, 6, 12, 18, 24],
                       labels=['Ночь (0-6)', 'Утро (6-12)', 'День (12-18)', 'Вечер (18-24)'])

heatmap_data = df.groupby(['DayPart', 'Hour']).agg({
    'cpu.usage.average': 'mean',
    'disk.usage.average': 'mean',
    'net.usage.average': 'mean'
}).unstack().fillna(0)

fig.add_trace(
    go.Heatmap(
        z=heatmap_data['cpu.usage.average'].values,
        x=heatmap_data.columns.get_level_values(1),
        y=heatmap_data.index,
        colorscale='Viridis',
        showscale=True,
        name='CPU по часам'
    ),
    row=4, col=1
)

# 11. Анализ паттернов нагрузки CPU
# Скользящее среднее
df['CPU_MA_3h'] = df['cpu.usage.average'].rolling(window=6, center=True).mean()  # 6 точек = 3 часа
df['CPU_MA_6h'] = df['cpu.usage.average'].rolling(window=12, center=True).mean()  # 12 точек = 6 часов

fig.add_trace(
    go.Scatter(
        x=df['Timestamp'],
        y=df['cpu.usage.average'],
        mode='lines',
        line=dict(color='lightgray', width=1),
        name='CPU Usage (факт)',
        showlegend=False
    ),
    row=4, col=2
)

fig.add_trace(
    go.Scatter(
        x=df['Timestamp'],
        y=df['CPU_MA_3h'],
        mode='lines',
        line=dict(color='blue', width=2),
        name='Скользящее среднее (3ч)'
    ),
    row=4, col=2
)

fig.add_trace(
    go.Scatter(
        x=df['Timestamp'],
        y=df['CPU_MA_6h'],
        mode='lines',
        line=dict(color='red', width=2),
        name='Скользящее среднее (6ч)'
    ),
    row=4, col=2
)

# Выделяем пиковые периоды
peak_periods = df[df['cpu.usage.average'] > df['cpu.usage.average'].quantile(0.90)]
fig.add_trace(
    go.Scatter(
        x=peak_periods['Timestamp'],
        y=peak_periods['cpu.usage.average'],
        mode='markers',
        marker=dict(color='red', size=8, symbol='diamond'),
        name='Пиковые нагрузки (>90%)'
    ),
    row=4, col=2
)

# 12. Сводный анализ эффективности
# Создаем таблицу с ключевыми показателями
summary_table = [
    ["Показатель", "Значение", "Статус", "Рекомендация"],
    ["Средний CPU Usage", f"{df['cpu.usage.average'].mean():.2f}%",
     "✅ Норма" if df['cpu.usage.average'].mean() < 50 else "⚠️ Высокий",
     "В норме" if df['cpu.usage.average'].mean() < 50 else "Рассмотреть оптимизацию"],

    ["Пиковый CPU Usage", f"{df['cpu.usage.average'].max():.2f}%",
     "⚠️ Высокий" if df['cpu.usage.average'].max() > 80 else "✅ Норма",
     "Мониторить пики" if df['cpu.usage.average'].max() > 80 else "В норме"],

    ["Средний CPU Ready", f"{df['cpu.ready.summation'].mean():.0f} ms",
     "⚠️ Высокий" if df['cpu.ready.summation'].mean() > 2000 else "✅ Норма",
     "Проверить хостовую нагрузку" if df['cpu.ready.summation'].mean() > 2000 else "В норме"],

    ["Средняя память", f"{df['mem.usage.average'].mean():.2f}%",
     "✅ Норма" if df['mem.usage.average'].mean() < 70 else "⚠️ Высокий",
     "В норме" if df['mem.usage.average'].mean() < 70 else "Увеличить память"],

    ["Пиковая сеть", f"{df['net.usage.average'].max():.0f} KB/s",
     "✅ Норма" if df['net.usage.average'].max() < 5000 else "⚠️ Высокий",
     "В норме" if df['net.usage.average'].max() < 5000 else "Проверить сетевую нагрузку"],

    ["Пиковый диск", f"{df['disk.usage.average'].max():.0f} KB/s",
     "✅ Норма" if df['disk.usage.average'].max() < 500 else "⚠️ Высокий",
     "В норме" if df['disk.usage.average'].max() < 500 else "Оптимизировать I/O"],

    ["Консистентность", f"{(df['cpu.usage.average'].std()/df['cpu.usage.average'].mean()*100):.1f}%",
     "✅ Стабильно" if (df['cpu.usage.average'].std()/df['cpu.usage.average'].mean()*100) < 100 else "⚠️ Нестабильно",
     "Нагрузка стабильна" if (df['cpu.usage.average'].std()/df['cpu.usage.average'].mean()*100) < 100 else "Исследовать колебания"]
]

fig.add_trace(
    go.Table(
        header=dict(
            values=summary_table[0],
            fill_color='lightblue',
            align='left',
            font=dict(size=12, color='black')
        ),
        cells=dict(
            values=list(zip(*summary_table[1:])),
            fill_color=[['lavender', 'white'] * 4],
            align='left',
            font=dict(size=11)
        )
    ),
    row=4, col=3
)

# Обновление макета
fig.update_layout(
    title_text='КОМПЛЕКСНЫЙ АНАЛИЗ МЕТРИК ВИРТУАЛЬНОГО СЕРВЕРА AIRFLOW1',
    height=1600,
    width=1800,
    showlegend=True,
    template='plotly_white'
)

# Обновление подписей осей
fig.update_xaxes(title_text="Время", row=1, col=1)
fig.update_yaxes(title_text="CPU Usage (%)", row=1, col=1)
fig.update_xaxes(title_text="Время", row=1, col=2)
fig.update_yaxes(title_text="CPU Ready (ms)", row=1, col=2)
fig.update_xaxes(title_text="Время", row=1, col=3)
fig.update_yaxes(title_text="CPU MHz", row=1, col=3)
fig.update_xaxes(title_text="Время", row=2, col=1)
fig.update_yaxes(title_text="Disk Usage (KB/s)", row=2, col=1)
fig.update_xaxes(title_text="Время", row=2, col=2)
fig.update_yaxes(title_text="Memory Usage (%)", row=2, col=2)
fig.update_xaxes(title_text="Время", row=2, col=3)
fig.update_yaxes(title_text="Network (KB/s)", row=2, col=3)
fig.update_xaxes(title_text="Время", row=3, col=1)
fig.update_yaxes(title_text="Memory (MB)", row=3, col=1)
fig.update_xaxes(title_text="Метрики", row=3, col=2)
fig.update_yaxes(title_text="Метрики", row=3, col=2)
fig.update_xaxes(title_text="Часы", row=4, col=1)
fig.update_yaxes(title_text="Время суток", row=4, col=1)
fig.update_xaxes(title_text="Время", row=4, col=2)
fig.update_yaxes(title_text="CPU Usage (%)", row=4, col=2)

fig.update_xaxes(tickformat='%H:%M', row=1, col=1)
fig.update_xaxes(tickformat='%H:%M', row=1, col=2)
fig.update_xaxes(tickformat='%H:%M', row=1, col=3)
fig.update_xaxes(tickformat='%H:%M', row=2, col=1)
fig.update_xaxes(tickformat='%H:%M', row=2, col=2)
fig.update_xaxes(tickformat='%H:%M', row=2, col=3)
fig.update_xaxes(tickformat='%H:%M', row=3, col=1)
fig.update_xaxes(tickformat='%H:%M', row=4, col=2)

# Сохраняем дашборд
fig.write_html("vm_metrics_analysis_airflow1.html")

print("\n" + "="*80)
print("ДЕТАЛЬНЫЙ АНАЛИЗ МЕТРИК")
print("="*80)

# Углубленный анализ
print("\n1. АНАЛИЗ CPU НАГРУЗКИ:")
print(f"   • Средняя загрузка CPU: {df['cpu.usage.average'].mean():.2f}%")
print(f"   • Максимальная загрузка CPU: {df['cpu.usage.average'].max():.2f}% в {df.loc[df['cpu.usage.average'].idxmax(), 'Timestamp']}")
print(f"   • Минимальная загрузка CPU: {df['cpu.usage.average'].min():.2f}%")
print(f"   • Стандартное отклонение: {df['cpu.usage.average'].std():.2f}%")

cpu_percentiles = df['cpu.usage.average'].quantile([0.25, 0.50, 0.75, 0.90, 0.95])
print(f"   • Процентили: 25%={cpu_percentiles[0.25]:.2f}%, 50%={cpu_percentiles[0.50]:.2f}%, "
      f"75%={cpu_percentiles[0.75]:.2f}%, 90%={cpu_percentiles[0.90]:.2f}%, 95%={cpu_percentiles[0.95]:.2f}%")

print("\n2. АНАЛИЗ CPU READY (КОНКУРЕНЦИЯ ЗА РЕСУРСЫ):")
print(f"   • Среднее CPU Ready: {df['cpu.ready.summation'].mean():.0f} ms")
print(f"   • Максимальное CPU Ready: {df['cpu.ready.summation'].max():.0f} ms")
print(f"   • Процент времени с CPU Ready > 2000ms: "
      f"{(len(df[df['cpu.ready.summation'] > 2000]) / len(df) * 100):.1f}%")

print("\n3. АНАЛИЗ ПАМЯТИ:")
print(f"   • Среднее использование памяти: {df['mem.usage.average'].mean():.2f}%")
print(f"   • Потребление памяти (среднее): {df['mem.consumed.average'].mean() / 1024 / 1024:.0f} MB")
print(f"   • Memory Ballooning (vmmemctl): {df['mem.vmmemctl.average'].mean():.2f}%")
print("     Примечание: 0% - ballooning не используется (хорошо)")

print("\n4. АНАЛИЗ ДИСКА И СЕТИ:")
print(f"   • Средняя нагрузка диска: {df['disk.usage.average'].mean():.0f} KB/s")
print(f"   • Пиковая нагрузка диска: {df['disk.usage.average'].max():.0f} KB/s")
print(f"   • Средняя сетевая нагрузка: {df['net.usage.average'].mean():.0f} KB/s")
print(f"   • Пиковая сетевая нагрузка: {df['net.usage.average'].max():.0f} KB/s")

print("\n5. ВРЕМЕННЫЕ ПАТТЕРНЫ:")
df['Hour'] = df['Timestamp'].dt.hour
hourly_stats = df.groupby('Hour').agg({
    'cpu.usage.average': 'mean',
    'disk.usage.average': 'mean',
    'net.usage.average': 'mean'
}).round(2)

print("\n   Средняя нагрузка по часам:")
print("   Час  | CPU(%) | Диск(KB/s) | Сеть(KB/s)")
print("   " + "-"*40)
for hour in range(24):
    if hour in hourly_stats.index:
        stats = hourly_stats.loc[hour]
        print(f"   {hour:2d}:00 | {stats['cpu.usage.average']:6.2f} | {stats['disk.usage.average']:10.0f} | {stats['net.usage.average']:10.0f}")

print("\n6. КОРРЕЛЯЦИОННЫЙ АНАЛИЗ:")
print("   Наиболее сильные корреляции:")
corr_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr_value = abs(correlation_matrix.iloc[i, j])
        if corr_value > 0.7:  # Сильная корреляция
            corr_pairs.append((correlation_matrix.columns[i],
                              correlation_matrix.columns[j],
                              correlation_matrix.iloc[i, j]))

corr_pairs.sort(key=lambda x: abs(x[2]), reverse=True)
for metric1, metric2, corr in corr_pairs[:5]:
    print(f"   • {metric1} ↔ {metric2}: {corr:.3f}")

print("\n" + "="*80)
print("ВЫВОДЫ И РЕКОМЕНДАЦИИ")
print("="*80)

# Формирование выводов
recommendations = []

# Анализ CPU
if df['cpu.usage.average'].mean() > 50:
    recommendations.append(("Высокая средняя загрузка CPU",
                          f"Средняя загрузка {df['cpu.usage.average'].mean():.1f}% > 50%",
                          "Рассмотреть увеличение CPU или оптимизацию приложения"))
else:
    recommendations.append(("CPU загрузка в норме",
                          f"Средняя загрузка {df['cpu.usage.average'].mean():.1f}%",
                          "Мониторить пиковые нагрузки"))

if df['cpu.ready.summation'].mean() > 2000:
    recommendations.append(("Высокий CPU Ready",
                          f"Среднее {df['cpu.ready.summation'].mean():.0f} ms > 2000 ms",
                          "Проверить конкуренцию на хосте, возможна нехватка CPU ресурсов"))
else:
    recommendations.append(("CPU Ready в норме",
                          f"Среднее {df['cpu.ready.summation'].mean():.0f} ms",
                          "Конкуренция за CPU в пределах нормы"))

# Анализ памяти
if df['mem.usage.average'].mean() > 70:
    recommendations.append(("Высокое использование памяти",
                          f"Среднее {df['mem.usage.average'].mean():.1f}% > 70%",
                          "Рассмотреть увеличение памяти"))
else:
    recommendations.append(("Память в норме",
                          f"Среднее использование {df['mem.usage.average'].mean():.1f}%",
                          "Достаточный объем памяти"))

# Анализ паттернов
peak_hours = hourly_stats[hourly_stats['cpu.usage.average'] > hourly_stats['cpu.usage.average'].quantile(0.75)]
if len(peak_hours) > 0:
    peak_hours_str = ', '.join([f"{int(h)}:00" for h in peak_hours.index])
    recommendations.append(("Выявлены часы пиковой нагрузки",
                          f"Пики в {peak_hours_str}",
                          "Планировать задачи вне пиковых часов"))

# Общие рекомендации
print("\nКЛЮЧЕВЫЕ РЕКОМЕНДАЦИИ:")
for i, (issue, data, recommendation) in enumerate(recommendations, 1):
    print(f"{i}. {issue}")
    print(f"   Данные: {data}")
    print(f"   Рекомендация: {recommendation}")
    print()

# Эффективность конфигурации
print("ЭФФЕКТИВНОСТЬ КОНФИГУРАЦИИ:")
print(f"   • Соотношение CPU Load/CPU Ready: {df['cpu.usage.average'].mean() / max(df['cpu.ready.summation'].mean() / 1000, 1):.2f}")
print(f"   • Эффективность памяти: {df['mem.usage.average'].mean() / 100:.3f}")
print(f"   • Стабильность нагрузки: {1 - (df['cpu.usage.average'].std() / df['cpu.usage.average'].mean()):.3f}")

# Расчет потенциальной оптимизации
print("\nРАСЧЕТ ПОТЕНЦИАЛЬНОЙ ОПТИМИЗАЦИИ:")
current_cpu = 16  # Из предыдущего анализа конфигураций
current_ram = 128  # Из предыдущего анализа конфигураций

# Расчет требуемых ресурсов на основе метрик
required_cpu = (df['cpu.usagemhz.average'].max() / 2000) * 1.2  # Предполагаем 2GHz ядра
required_ram = (df['mem.consumed.average'].max() / 1024 / 1024 / 1024) * 1.2  # GB с запасом

print(f"   • Текущая конфигурация: {current_cpu} CPU, {current_ram} GB RAM")
print(f"   • Рекомендуемая конфигурация: {required_cpu:.1f} CPU, {required_ram:.1f} GB RAM")
print(f"   • Возможная экономия CPU: {max(0, current_cpu - required_cpu):.1f} ядер")
print(f"   • Возможная экономия RAM: {max(0, current_ram - required_ram):.1f} GB")

print("\n" + "="*80)
print(f"✅ Детальный анализ сохранен как 'vm_metrics_analysis_airflow1.html'")
print("Откройте файл в браузере для просмотра интерактивных графиков")
print("="*80)

# Экспорт ключевых метрик в CSV для дальнейшего анализа
key_metrics = df[['Timestamp', 'cpu.usage.average', 'cpu.ready.summation',
                  'mem.usage.average', 'disk.usage.average', 'net.usage.average']].copy()
key_metrics.to_csv('airflow1_key_metrics.csv', index=False)
print(f"✅ Ключевые метрики сохранены в 'airflow1_key_metrics.csv'")
```
